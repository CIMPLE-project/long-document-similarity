# Data

```python
df_annotated_claims = pd.read_csv('/data/peskine/long_document_similarity/dataset0.csv')
df_annotated_claims
```

```python
df_claims = pd.read_csv('/data/peskine/birdwatch/birdwatch_match_dataset.csv')
df_claims
# this dataset matches tweets and claims together
# some tweets are not matched to any claims
# multiple tweets can be matched to the same claim
```

```python
all_claims = df_claims['claims'].tolist()
all_claims = list(set(all_claims))[1:]
len(all_claims)
```

```python
all_claims = df_claims['claims'].tolist()
all_claims = list(set(all_claims))[1:]

all_claims += df_annotated_claims['genericClaim'].tolist()
all_claims += df_annotated_claims['specificClaim'].tolist()
all_claims += df_annotated_claims['hybridClaim'].tolist()
len(all_claims)
```

```python
all_articles = df_annotated_claims['articleText'].tolist()

all_chunks = []
all_article_ids = []
all_chunk_ids = []
chunk_number=0

for i in range(0, len(all_articles)):
    art = all_articles[i]
    chunks = art.split('\n')
    for c in chunks:
        if c != '' and c not in all_chunks:
            all_chunks.append(c)
            all_article_ids.append(int(i/2))
            all_chunk_ids.append(chunk_number)
            chunk_number+=1
            
```

```python
df_chunks = pd.DataFrame()
df_chunks['articleId'] =  all_article_ids
df_chunks['chunkId'] = all_chunk_ids
df_chunks['chunk'] = all_chunks
df_chunks.to_csv('/data/peskine/long_document_similarity/dataset_articles0.csv', index=False)
```

```python
article_ids = []
gens_youri = []
spes_youri = []
hybs_youri = []
gens_kata = []
spes_kata = []
hybs_kata = []
chunk_ids1 = [] 
chunk_ids2 = [] 

for i in range(0, 20):
    article_ids.append(i)
    
    chunk_ids1.append(df_annotated_claims['chunkId'].tolist()[i*2+0])
    chunk_ids2.append(df_annotated_claims['chunkId'].tolist()[i*2+1])

    
    gens_youri.append(all_claims.index(df_annotated_claims['genericClaim'].tolist()[i*2+0]))
    gens_kata.append(all_claims.index(df_annotated_claims['genericClaim'].tolist()[i*2+1]))

    spes_youri.append(all_claims.index(df_annotated_claims['specificClaim'].tolist()[i*2+0]))
    spes_kata.append(all_claims.index(df_annotated_claims['specificClaim'].tolist()[i*2+1]))

    hybs_youri.append(all_claims.index(df_annotated_claims['hybridClaim'].tolist()[i*2+0]))
    hybs_kata.append(all_claims.index(df_annotated_claims['hybridClaim'].tolist()[i*2+1]))

df_match = pd.DataFrame()
df_match['articleId'] = article_ids
df_match['chunkId1'] = chunk_ids1
df_match['genericClaim1'] = gens_youri
df_match['specificClaim1'] = spes_youri
df_match['hybridClaim1'] = hybs_youri

df_match['chunkId2'] = chunk_ids2
df_match['genericClaim2'] = gens_kata
df_match['specificClaim2'] = spes_kata
df_match['hybridClaim2'] = hybs_kata
df_match.to_csv('/data/peskine/long_document_similarity/dataset_matching0.csv', index=False)
```


# Model

```python
from sentence_transformers import SentenceTransformer, util
SBmodel = SentenceTransformer('all-mpnet-base-v2') # best average performance model
```

```python
#embedding tweets
emb_claims = SBmodel.encode(all_claims,show_progress_bar=True, batch_size=128)
#embedding claims
emb_chunks = SBmodel.encode(all_chunks,show_progress_bar=True, batch_size=128)
```

```python
cos_sim = util.cos_sim(emb_chunks, emb_claims)
```

# Matching

## Average

```python
ranks1 = []
ranks2 = []
for article_id in range (0, 20):
    chunk_ids = df_chunks[df_chunks['articleId']==article_id]['chunkId'].tolist()
    generic_target1 = df_match[df_match['articleId']==article_id]['hybridClaim1'].tolist()[0]
    generic_target2 = df_match[df_match['articleId']==article_id]['hybridClaim2'].tolist()[0]
    
    claims_similarity_scores = cos_sim[chunk_ids].sum(dim=0).numpy()
        
    claim_ids_sorted = claims_similarity_scores.argsort()[::-1].tolist()
    ranks1.append(claim_ids_sorted.index(generic_target1))
    ranks2.append(claim_ids_sorted.index(generic_target2))
```

```python
ranks1, ranks2, sum(ranks1)/20, sum(ranks2)/20
```

## Maximum

```python
ranks1 = []
ranks2 = []
for article_id in range (0, 20):
    chunk_ids = df_chunks[df_chunks['articleId']==article_id]['chunkId'].tolist()
    generic_target1 = df_match[df_match['articleId']==article_id]['hybridClaim1'].tolist()[0]
    generic_target2 = df_match[df_match['articleId']==article_id]['hybridClaim2'].tolist()[0]
    
    claims_similarity_scores = cos_sim[chunk_ids].max(dim=0)[0].numpy()
        
    claim_ids_sorted = claims_similarity_scores.argsort()[::-1].tolist()
    ranks1.append(claim_ids_sorted.index(generic_target1))
    ranks2.append(claim_ids_sorted.index(generic_target2))
    
```

```python
ranks1, ranks2, sum(ranks1)/20, sum(ranks2)/20
```

## Maximum specific chunk

```python
ranks1 = []
ranks2 = []
for article_id in range (0, 20):
    chunk_id1 = df_match[df_match['articleId']==article_id]['chunkId1'].tolist()[0]
    chunk_id2 = df_match[df_match['articleId']==article_id]['chunkId2'].tolist()[0]
    generic_target1 = df_match[df_match['articleId']==article_id]['specificClaim1'].tolist()[0]
    generic_target2 = df_match[df_match['articleId']==article_id]['specificClaim2'].tolist()[0]
    
    claims_similarity_scores = cos_sim[chunk_id1].numpy()
    claim_ids_sorted = claims_similarity_scores.argsort()[::-1].tolist()
    ranks1.append(claim_ids_sorted.index(generic_target1))
    
    claims_similarity_scores = cos_sim[chunk_id2].numpy()
    claim_ids_sorted = claims_similarity_scores.argsort()[::-1].tolist()

    ranks2.append(claim_ids_sorted.index(generic_target2))
    
```

```python
ranks1, ranks2, sum(ranks1)/20, sum(ranks2)/20
```
